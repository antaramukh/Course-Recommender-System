4.1 What is Content Based Recommendation
It is a type of recommendation system in which recommended movies have similar content with the already liked movies. But first, we need to define the content. Content can be a story, or script. We can take brief summary of the movie, cast, and genres the content of the movie, and recommend movies with similar content.

```python
df['content'] = df['course_title'] + ' ' + df['tags']
```

So, we will merge all of them in one text. But computers cannot understand the meaning of these texts, thus we need to apply some pre-processing operations to our "content". We will be using NLP (Natural Language Processing)

Natural language processing is a subfield of linguistics, computer science, and artificial intelligence concerned with the interactions between computers and human language, in particular how to program computers to process and analyze large amounts of natural language data.



We will be using a three-step process to transform our "content" into a form that could be understood by a computer algorithm and with which it can extract meaningful insights.

7.2 Tokenization
Tokenization is the process of breaking down sentence or paragraphs into smaller chunks of words called tokens.

7.3 Stop Words Removal
On removal of some words, the meaning of the sentence doesn't change, like and, am. Those words are called stop-words and should be removed before feeding to any algorithm. In datasets, some non-stop words repeat very frequently. Those words too should be removed to get an unbiased result from the algorithm.

```python
# Clean Text : Stopwords, Special Characters
df['course_title']=df['course_title'].str.lower()

df['content']=df['content'].str.lower()
df['content'] = df['content'].apply(nfx.remove_stopwords)
df['content'] = df['content'].apply(nfx.remove_special_characters)

df
```

```python
df=df.reset_index()
df=df[['course_title','url','tags','content']]
df.tail(25)
```

7.4 Vectorization
After tokenization, and stop words removal, our "content" are still in string format. We need to convert those strings to numbers based on their importance (features). We use TF-IDF vectorization to convert those text to vector of importance. With TF-IDF we can extract important words in our data. It assign rarely occurring words a high number, and frequently occurring words a very low number.

With Tfidfvectorizer on the contrary, you will do all three steps at once. Under the hood, it computes the word counts, IDF values, and Tf-idf scores all using the same dataset.


8.1 What is Cosine Similarity
After using generating TF-IDF vectors, we will get a vector for each movie "content". As they are vector, they can be visualized on a cartesian plane.
The above dataframe shows cosine similarity of all movies. Cosine similarity of 1 means that movie is very similar, and 0 means that movies are not similar.

To find similar movies, we just need to calculate angle between two vectors. If angle is small, then we can say that those movies are similar.



You might think that we can't place all movies in this small cartesian plane. Correct, we can't. But the TF-IDF vectors generally have 500+ dimensions. Now, think how many movies can be placed in 500+ dimension cartesian plane.

```python
vectorizer = TfidfVectorizer()
course_vectors = vectorizer.fit_transform(df['content'].values)
similarity = cosine_similarity(course_vectors)
```

```python
def recommend(title,num_of_rec=10):
    title=title.lower()
    # find movie index from dataset
    try:
        course_index = df[df['course_title'] == title].index[0]

        # finding cosine similarities of movie
        scores = list(enumerate(similarity[course_index]))

        # sorting cosine similarities
        sorted_scores = sorted(scores,key=lambda x:x[1],reverse=True)

        selected_course_indices = [i[0] for i in sorted_scores[1:]]
        selected_course_scores = [i[1] for i in sorted_scores[1:]]
        result = df[['course_title','url']].iloc[selected_course_indices]
        rec_df = pd.DataFrame(result)
        rec_df['similarity_scores'] = selected_course_scores
        rec_df['course_title']=rec_df['course_title'].str.title()
        return rec_df.head(num_of_rec)
    except:
        course_index = df[df[''] == title].index[0]

        # finding cosine similarities of movie
        scores = list(enumerate(similarity[course_index]))

        # sorting cosine similarities
        sorted_scores = sorted(scores,key=lambda x:x[1],reverse=True)

        selected_course_indices = [i[0] for i in sorted_scores[1:]]
        selected_course_scores = [i[1] for i in sorted_scores[1:]]
        result = df[['course_title','url']].iloc[selected_course_indices]
        rec_df = pd.DataFrame(result)
        rec_df['similarity_scores'] = selected_course_scores
        rec_df['course_title']=rec_df['course_title'].str.title()
        return rec_df.head(num_of_rec)
```

```python
recommend('Spatial Data Science And Applications',20)
```
